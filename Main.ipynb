## Training Model : Vectorization

```python
def Hypothesis(X, theta):
    
    return np.dot(X, theta)
```

```python
def Error(X, Y, theta):
    
    Y_predict = Hypothesis(X, theta)
    m = X.shape[0]
    error = np.sum((Y_predict - Y)**2)
    
    return error/m
```

```python
def Gradient(X, Y, theta):
    
    m = X.shape[0]
    Y_predict = Hypothesis(X, theta)
    grad = np.dot(X.T, (Y_predict - Y))
    return grad/m
```

```python

def GradientDescent(X, Y, LR = 0.1, maxiter = 300):
    n = X.shape[1]
    error_list = []
    theta = np.zeros((n,))
    
    for i in range(maxiter):
        error_list.append(Error(X, Y, theta))        
        theta = theta - Gradient(X,Y, theta) * LR
    
    return theta, error_list
```

```python
theta, error_list = GradientDescent(X, Y)
```

```python
plt.style.use("seaborn")
plt.plot(error_list);
```

```python
Y_predict = Hypothesis(X, theta)
```

$ R^2 score$

```python
def R2score(Y, Y_prediction):
    
    Numerator = np.sum((Y - Y_prediction) ** 2)
    Denominator = np.sum((Y - np.mean(Y_prediction)) ** 2)
    score = 1 - (Numerator/Denominator)
    return score
```

```python
R2score(Y, Y_predict)
```
